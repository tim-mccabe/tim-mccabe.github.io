```python
import pandas as pd
```

```python
data_file = "course_info_w_lat_lon.csv"
```

```python
data_file_pd = pd.read_csv(data_file)
data_file_pd.head()
```

```python
course_info = data_file_pd.drop(["Latitude", "Longitude"], axis=1)
course_info.head()
```

```python
course_info.to_csv("course_info.csv", index=False, header=True)
```

```python
attribute_file = "course_attributes.csv"
```

```python
attribute_df = pd.read_csv(attribute_file)
attribute_df.head()
```

```python
attribute_df = attribute_df.drop(["Total Score"], axis=1)
attribute_df.head()
```

```python
total_score = course_info.drop(["Course_Name", "Yards", "Par", "Location", "Architect", "Year_Built", "Restoration_Architect", "Year_Restored"], axis=1)
total_score.head()
```

```python
df_merge = pd.merge(attribute_df, total_score, on='Rank')

df_merge.head()
```

```python
renamed_df = df_merge.rename(columns={"Points": "Total Score"})
renamed_df.head()
```

```python
renamed_df.to_csv("course_attributes_new.csv", index=False, header=True)
```
